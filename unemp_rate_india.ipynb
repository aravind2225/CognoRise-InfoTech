```python
#importig libraries and reading data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv('/content/Unemp_india.csv')
df
```

```python
#finding the data types
df.dtypes
```

```python
#finding the dimensions and info
print(df.shape)
print(df.size)
df.info()
```

```python
print(df['Region'].value_counts())
print(df['Region'].nunique())
```

```python
#mining insights on Andhra Pradesh
df[df['Region']=='Andhra Pradesh']
```

```python
#Assigning the new column names (Same Names but removing the spaces)
df.columns=['Region','Date','Frequency','Estimated Unemployment Rate (%)','Estimated Employed','Estimated Labour Participation Rate (%)','Area']
df.info()
```

```python
#Changing the date column dtypes from string the date
df['Date']=df['Date'].astype('datetime64[ns]')
df.info()
```

```python
#Adding the new column 'Year' for before and after corona analysis
df['Year'] = df['Date'].dt.year
df['Year']=df['Year'].astype('object')
df
```

```python
#finding the unemployment rate state wise
df.groupby('Region')['Estimated Unemployment Rate (%)'].mean()
```

```python
#finding the unemployment rate by state wise before and after corona
unemp_rate_change=df.groupby(['Region','Year'])['Estimated Unemployment Rate (%)'].mean().reset_index()
unemp_rate_change.columns=['State','Year','Unemployment Rate']
```

```python
unemp_rate_change
```

```python
#plotting the unemployment change rate before and after corona
pivot_df = unemp_rate_change.pivot(index='State', columns='Year', values='Unemployment Rate')
pivot_df.plot(kind='bar', figsize=(10, 6), width=0.8)
plt.title('Clustered Column Chart of Unemployment Rate by Region and Year')
plt.xlabel('State')
plt.ylabel('Unemployment Rate (%)')
plt.legend(title='Year')
plt.show()
```

```python
#converting normal number to the millions for estimated employed count
df['Estimated Employed']=df['Estimated Employed'].astype('float64')
df['Estimated Employed']=df['Estimated Employed']/1000000
df
```

```python
#extracting total number of employed people(in millions) before and after corona
employed_count=df.groupby(['Region','Year'])['Estimated Employed'].mean().reset_index()
employed_count.columns=['State','Year','Total Employed(in Million)']
employed_count
```

```python
#plotting the above table as clusterd column chart
employed_count_pivot=pd.pivot(employed_count,index='State',columns='Year',values='Total Employed(in Million)')
employed_count_pivot.plot(kind='bar', figsize=(10, 6), width=0.8)
```

```python
#finding the avg unemp rate in rural and urban
df.groupby('Area')['Estimated Unemployment Rate (%)'].mean()
```

```python
#Now finding the average Participated labour force in each state overall(before and after corona combinedly)
df.groupby('Region')['Estimated Labour Participation Rate (%)'].mean().reset_index()
```

```python
#labour participation rate before and corona for each state
labour_prate=df.groupby(['Region','Year'])['Estimated Labour Participation Rate (%)'].mean().reset_index()
labour_prate.columns=['State','Year','Labour Rate']
labour_prate_pivot=pd.pivot(labour_prate,index='State',columns='Year',values='Labour Rate')
labour_prate_pivot.plot(kind='bar', figsize=(12, 7), width=0.8)
```

```python
#Trend analysis according the date and year for differnet states about unemployment rate
trend_analsys=df.groupby(['Region','Date'])['Estimated Unemployment Rate (%)'].mean().reset_index()
trend_analsys.columns=['State','Date','Unemp_rate']
trend_analsys_pivot=pd.pivot(trend_analsys,index='Date',columns='State',values='Unemp_rate')
trend_analsys_pivot.plot(kind='line',figsize=(20, 10),legend='State')
```
